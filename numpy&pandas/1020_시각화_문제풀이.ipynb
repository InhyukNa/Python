```python
x = iris.petal_length.values
# type(x)
# x.shape
sns.rugplot(x)
plt.title('iris 데이터 중, 꽃잎의 길이에 대한 Rug Plot')
```

```python
sns.kdeplot(x)
plt.title('iris 데이터 중, 꽃잎의 길이에 대한 Kernel Density Plot')
```

```python
sns.distplot(x, kde=True, rug =True)
plt.title('iris 데이터 중 꽃잎의 길이에 대한 Dist Plot')
```

```python
sns.countplot(x ='class', data=titanic)
plt.title('타이타닉호의 각 클래스별 승객 수')
```

```python
sns.countplot(x='day', data=tips)
plt.title('요일별 팁을 준 횟수')
```

```python
sns.jointplot(x='sepal_length', y = 'sepal_width' , data=iris)
plt.title('꽃받침의 길이와 넓이의 Join Plot')
```

```python
sns.jointplot(x="sepal_length", y='sepal_width', data = iris, kind ='kde')
plt.title('꽃받침의 길이와 넓이의 Jont Plot 과 Kernel Density Plot')
```

```python
sns.pairplot(iris)
plt.title("Iris Data의 Pair Plot")
```

```python
titanic_size = titanic.pivot_table(
 index='class', columns='sex', aggfunc='size'
)
titanic_size
```

```python
'''
- survived : 생존여부(1:생존, 0:사망)
- pclass : 승선권 클래스(1:1st, 2:2nd, 3:3rd)
- sex : 승객 성별
- age : 승객 나이
- sibsp : 동반한 형제자매, 배우자 수
- parch : 동반한 부모, 자식 수
- fare : 티켓의 요금
- embarked :승선한 항구명(C:cherbourg, Q:Queenstown, S:Southampton)
- class :
- who : man, woman, child
'''
```

#### 문제1. 모든 속성들이 생존자와 관계가 있는지 산포도 를 그려본다.

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset('titanic') # 타이타닉호 데이터 로딩
# type(titanic)
sns.pairplot(titanic, hue='survived', markers=['o','s'])
plt.title('모든 속성을 생존자로 시각화')
```

###  2개 속성간의 통계치(countplot)를 구해보자.

#### 문제2. 여성과 남성의 생존률의 차이가 있는가?

```python
sns.countplot('sex', hue='survived', data=titanic)
```

#### 문제3. 승선권 클래스 별로 생존률의 차이가 있는가?

```python
sns.countplot('pclass', hue='survived', data=titanic)
```

#### 문제4. 혼자인 사람보다 형제나 배우자가 있는 사람이 생존률이 높은가?

```python
sns.countplot('sibsp', hue='survived', data=titanic)
```

#### 문제5. 부모나 자식이 있는 사람이 생존률이 높은가?

```python
sns.countplot('parch', hue='survived', data=titanic)
```

#### 문제6. 승선한 항구명에 따라 생존률의 차이가 있는가?

```python
sns.countplot('embarked', hue='survived', data=titanic)
```

#### 문제8. Heapmap을 사용하여 승선권 클래스와 WHO(남자, 여자, 아이)의 수를 표현 하시오.

```python
titanic_size = titanic.pivot_table(
 index='pclass', columns='who', aggfunc='size'
)
titanic_size
```

```python
sns.heatmap(titanic_size, cmap = sns.light_palette("green", as_cmap=True),
           annot=True, fmt ="f")
plt.title('Heatmap')
```

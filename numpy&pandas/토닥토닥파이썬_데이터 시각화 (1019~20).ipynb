```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.displot(tips['total_bill'], bins = 40)  #displot : displayplot
plt.show()
```

```python
# 두가지 속성을 비교하여 rendering, scatter, distribute
sns.jointplot(x = 'total_bill', y = 'tip', data = tips)
plt.show()
```

```python
# 전체 속성들을 한 눈에 보고싶을때
sns.pairplot(tips, hue = 'sex')  # 성별 분포를 쉽게 보여준다
plt.show()
```

```python
# 카테고리 데이터를 구분해서 얼마씩 차이나는지를 보여준다. ex) 남,여/흡연,비흡연
sns.barplot(x = 'sex', y = 'total_bill', data = tips)
plt.show()
```

```python
# 분류된 카테고리 기준에 따라 갯수 헤아리기
sns.countplot(x = 'sex', data = tips)
plt.show()
```

```python
sns.barplot(x = 'day', y = 'total_bill', data = tips)
plt.show()
```

```python
# 값의 분포 보기
sns.boxplot(x = 'day', y = 'total_bill', data = tips)
plt.show()
```

```python
# 격자형 차트 유형
# 상관관계를 격자형태로 보여준다
# 숫자들의 level에 따라 색이 다르게 구분된다
tc = tips.corr()
sns.heatmap(tc, annot = True)
tips
```

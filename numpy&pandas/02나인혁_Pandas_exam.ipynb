```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### 문제1)  남자(male) 이면서 Pclass 가 1인 승객은 몇 명인가요?

```python
cnt = titanic.loc[(titanic['Sex'] == 'male') & (titanic['Pclass'] == 1), 'PassengerId'].count()
cnt
```

####  문제2) 남자이면서 Pclass 가 1이고 Fare가 50 이상인 승객은 몇 명인가요?

```python
cntm = titanic.loc[(titanic['Sex'] == 'male') & (titanic['Pclass'] == 1) & (titanic['Fare'] >= 50), 'PassengerId'].count()
cntm
```

#### 문제3) 새로운 열(sex_code)을 만들어서 남자인 경우는 코드 0, 여자인 경우는 1로 저장

```python
sex_code = pd.DataFrame([0, 1], index = ['male', 'female'], columns = ['sex_code'])
sex_code
```

```python
titanic2 = pd.merge(titanic, sex_code, left_on = 'Sex', right_index = True, how = 'inner')
titanic2
```

#### 문제4) 성별 생존자 수의 합(남자 생존자 수의 합, 여자 생존자수의 합) - 생존자는 Survived 의 값이 1인 것.

```python
sv = titanic2.groupby('Sex')['Survived'].sum()
```

#### 문제7) 각 영화 별 평균 평점을 movieid, title, 평균 평점 순으로 출력 하세요.

```python
movie_mean = df_r.groupby('movieId')['rating'].mean()
result_df = movie_mean.to_frame()  
result_df.columns = ['average'] 
df = pd.merge(df_m, result_df, left_on = 'movieId', right_index = True, how = 'inner')
df[['movieId', 'title', 'average']]
```

#### 문제8) 평균 평점이 가장 높은 영화의 제목을 출력(동률인 경우 모두 출력) 하세요.

```python
ch = df.loc[df['average'] == df['average'].max()]
ch
```

```python
print(ch['title'])
```

#### 문제9) Comedy 영화 중 가장 평점이 낮은 영화 제목 출력(동률이 있는 경우 모두 출력) 하세요.

```python
comedy = df.loc[df['genres'].str.contains('Comedy')]
cl = comedy.loc[comedy['average'] == comedy['average'].min()]
cl
```

```python
print(cl['title'])
```

#### 문제10) 2015년도에 평가된 모든 Romance 영화의 평균 평점을 구하세요.

```python
import time
```

```python
df_2 = pd.merge(df_m, df_r, left_on = 'movieId', right_on = 'movieId', how = 'inner')
result = df_2.loc[df_2['genres'].str.contains('Romance')]

start = time.mktime(time.strptime('2015-01-01 00:00:00', '%Y-%m-%d %H:%M:%S'))
end = time.mktime(time.strptime('2016-01-01 00:00:00', '%Y-%m-%d %H:%M:%S'))
print(result.loc[(result['timestamp'] >= start) & (result['timestamp'] < end), 'rating'].mean())
```

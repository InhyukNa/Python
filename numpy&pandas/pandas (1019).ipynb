### 2. 각 사용자별 평균 평점을 출력하세요.

```python
df_rating.groupby('userId')['rating'].mean()
```

### 3. 각 영화 별 평균 평점을 movieid, title, 평균평점 순으로 출력하세요

```python
movie_mean = df_rating.groupby('movieId')['rating'].mean()
result_df = movie_mean.to_frame()  # Series -> DataFrame 변환
result_df.columns = ['average']  # 집계해서 새로 생긴 컬럼의 이름을 average로 해준다
df = pd.merge(df_movie, result_df, left_on = 'movieId', right_index = True, how = 'inner')
df[['movieId', 'title', 'average']]
```

```python
df[['movieId', 'title', 'average']].sort_values('average')
```

```python
df[['movieId', 'title', 'average']].sort_values('average', ascending = False)
```

### 4. 평균평점이 가장 높은 영화의 제목을 출력하세요 (동률이 있는 경우 모두 출력하세요)

```python
df.loc[df['average'] == df['average'].max()]
```

```python
df.loc[df['average'] == df['average'].max(), 'title']
```

### 5. Comedy 영화 중 가장 평점이 낮은 영화 제목 출력 (동률이 있는 경우 모두 출력하세요)

```python
comedyDf = df.loc[df['genres'].str.contains('Comedy')]
comedyDf.loc[comedyDf['average'] == comedyDf['average'].min()]
```

### 6. 2015년도에 평가된 모든 Romance 영화의 평균 평점

```python
import time
```

```python
df2 = pd.merge(df_movie, df_rating, left_on = 'movieId', right_on = 'movieId', how = 'inner')
result = df2.loc[df2['genres'].str.contains('Romance')]

result_df = pd.DataFrame(columns = ['평점'])
start = time.mktime(time.strptime('2015-01-01 00:00:00', '%Y-%m-%d %H:%M:%S'))
end = time.mktime(time.strptime('2016-01-01 00:00:00', '%Y-%m-%d %H:%M:%S'))
print(result.loc[(result['timestamp'] >= start) & (result['timestamp'] < end), 'rating'].mean())
```

### Pivot Table

```python
df_phone = pd.read_csv('phone_data.csv')
```

```python
df_phone
```

```python
df_phone.pivot_table(['duration'],
                    index = [df_phone.month, df_phone.item],
                    columns = df_phone.network, aggfunc = 'sum', fill_value = 0)
```

```python
df_movie = pd.read_csv('movie_rating.csv')
df_movie.head()
```

```python
df_movie.pivot_table(['rating'], index = df_movie.critic, columns = df_movie.title, 
                    aggfunc = 'sum', fill_value = 0)
```

#### crosstab

```python
pd.crosstab(index = df_movie.critic, columns = df_movie.title, values = df_movie.rating,
           aggfunc = 'first').fillna(0)
```

```python
df_movie.groupby(['critic', 'title']).agg({'rating':'first'}).unstack().fillna(0)
```

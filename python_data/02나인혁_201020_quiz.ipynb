```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.linspace(0.0, 5.0, 6)
x
```

```python
x = np.linspace(0,5,6)
y_n = 2 ** x
y_nlogn = x * y_n
y_n2 = y_n ** 2
y_n3 = y_n ** 3
y_2n = 2 ** y_n
print(x)
print(y_n)
print(y_nlogn)
print(y_n2)
print(y_n3)
print(y_2n)
```

```python
plt.plot(x, y_n)
plt.show()
```

```python
plt.plot(x, y_nlogn)
plt.show()
```

```python
plt.plot(x, y_n2)
plt.show()
```

```python
plt.plot(x, y_n3)
plt.show()
```

```python
plt.plot(x, y_2n)
plt.show()
```

```python
plt.subplot(2,3,1)
plt.plot(x, y_n)

plt.subplot(2,3,2)
plt.plot(x, y_nlogn)

plt.subplot(2,3,3)
plt.plot(x, y_n2)

plt.subplot(2,3,4)
plt.plot(x, y_n3)

plt.subplot(2,3,5)
plt.plot(x, y_2n)

plt.show()
```

```python
plt.subplot(2,3,1)
plt.plot(x, y_n, color = 'red', linewidth = 1, alpha = 0.7,
       linestyle = '--', marker = 'o', markersize = 5,
       markerfacecolor = 'violet', markeredgewidth = 3, markeredgecolor = 'hotpink')

plt.subplot(2,3,2)
plt.plot(x, y_nlogn, color = 'green', linewidth = 1, alpha = 0.7,
       linestyle = '--', marker = '*', markersize = 5,
       markerfacecolor = 'green', markeredgewidth = 3, markeredgecolor = 'green')

plt.subplot(2,3,3)
plt.plot(x, y_n2, color = 'blue', linewidth = 1, alpha = 0.7,
       linestyle = '--', marker = 's', markersize = 5,
       markerfacecolor = 'skyblue', markeredgewidth = 3, markeredgecolor = 'skyblue')

plt.subplot(2,3,4)
plt.plot(x, y_n3,  color = 'brown', linewidth = 1, alpha = 0.7,
       linestyle = '--', marker = '^', markersize = 5,
       markerfacecolor = 'brown', markeredgewidth = 3, markeredgecolor = 'brown')

plt.subplot(2,3,5)
plt.plot(x, y_2n, color = 'black', linewidth = 1, alpha = 0.7,
       linestyle = '--', marker = 'h', markersize = 5,
       markerfacecolor = 'gray', markeredgewidth = 3, markeredgecolor = 'gray')

plt.show()
```

```python
x = np.arange(6)
y = x ** 2
logn = np.log2(x)
nlogn = x * logn
x_3 = x ** 3
z = 2 ** x
print(x)
print(y)
print(logn)
print(nlogn)
print(x_3)
print(z)
```
